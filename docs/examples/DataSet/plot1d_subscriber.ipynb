```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import time 

from qcodes.dataset.live_monitor import Plot1DSubscriber
from qcodes.dataset.data_set import new_data_set, ParamSpec
```

```python
data_set = new_data_set(
    "test", 
    specs=[
        ParamSpec("x", "numeric", unit='Hz'), 
        ParamSpec("y", "numeric", unit='V'), 
        ParamSpec("z", "numeric", unit='I')
    ]
)
```

```python
data_set.subscribe(Plot1DSubscriber({"x name": 0, "y name": 1}))
data_set.subscribe(Plot1DSubscriber({"x name": 0, "z name": 2}))
```

```python
for x in np.linspace(100, 200, 150):
    y = np.random.randn()
    z = -y
    data_set.add_result({"x": x, "y": y, "z": z})
    time.sleep(0.1)
```
